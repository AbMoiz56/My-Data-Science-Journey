```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Netflix.csv")
data
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data[data.columns[data.isnull().any()]].isnull().sum
```

```python
data['director'] = data['director'].fillna('Unknown')
data['cast'] = data['cast'].fillna('Unknown')
data['country'] = data['country'].fillna('Unknown')
```

```python
data['date_added'] = data['date_added'].fillna(data['date_added'].mode()[0])
data['rating'] = data['rating'].fillna(data['rating'].mode()[0])
data['duration'] = data['duration'].fillna(data['duration'].mode()[0])
```

```python
data.duplicated().sum()
```

```python
data.describe()
```

```python
sns.histplot(data['release_year'],kde=True,bins=30)
plt.title("Distribution of release year")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.show()
```

```python
data['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen'])
plt.title("Content Type Distribution")
plt.ylabel("") 
plt.show()
```

```python
sns.boxplot(x=data['release_year'])
plt.title("Box plot of release year")
plt.show()
```

```python
data['rating'].value_counts().plot(kind='bar', figsize=(10,4))
plt.title("Count of Shows by Rating")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(16, 6))  
sns.countplot(data=data, x='release_year', hue='type')
plt.xticks(rotation=90)
plt.title("Movies vs TV Shows by Release Year")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
from collections import Counter

genre_list = data['listed_in'].str.split(', ')
flat_list = [genre for sublist in genre_list.dropna() for genre in sublist]
top_genres = Counter(flat_list).most_common(10)
genre_df = pd.DataFrame(top_genres, columns=['Genre', 'Count'])

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(data=genre_df, x='Count', y='Genre')
plt.title("Top 10 Most Frequent Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()
```

```python
top_directors = data['director'].value_counts().head(10)
top_directors


```

```python
top_countries = data['country'].value_counts().head(10).index
filtered = data[data['country'].isin(top_countries)]

plt.figure(figsize=(15,6))
sns.countplot(data=filtered, x='country', hue='type')
plt.title("Top Countries by Type of Content")
plt.ylabel("Number of Titles")
plt.xlabel("Country")
plt.show()
```

```python
data['date_added'] = pd.to_datetime(data['date_added'])

data['month_added'] = data['date_added'].dt.month_name()


plt.figure(figsize=(12,6))
sns.countplot(data=data, x='month_added', order=[
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
])
plt.title("Content Added by Month")
plt.xticks(rotation=45)
plt.ylabel("Number of Titles")
plt.xlabel("Month")
plt.show()
```
